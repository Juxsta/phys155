```python
from scipy import signal
import matplotlib.pyplot as plt
f = 200
N = math.ceil(40*f)
t = np.linspace(0, 1/f*10, N, endpoint=False).astype(np.float128)

sig = signal.square(2 * np.pi * f * t)
plt.plot(t, sig)
plt.ylim(-2, 2)
```

```python
sos = signal.butter(1, 10, 'low', fs=N, output='sos')
filtered_sig = np.nan_to_num(signal.sosfilt(sos, sig)).astype(np.float)
plt.plot(t,filtered_sig)
```
